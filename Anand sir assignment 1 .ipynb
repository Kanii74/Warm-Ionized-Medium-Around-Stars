```python
import numpy as np
import matplotlib.pyplot as plt

# constants
h = 6.62607015e-34
c = 299792458.0
k = 1.380649e-23

# Function to calculate Planck's Law
def planck_function(frequency, temperature):
    return ((2 * h * frequency**3) / ((c**2) * (np.exp((h * frequency) / (k * temperature)) - 1)))

# Spectral types and their corresponding temperatures (in Kelvin)
spectral_types = ['O', 'B', 'A', 'F', 'G', 'K', 'M']
temperatures = [30000, 25000, 10000, 7000, 5500, 4000, 3000]

# Frequency range (in Hz)
frequency_range = np.logspace(13, 20, 10000)

# Plotting the synthetic Planck curves
plt.figure(figsize=(10, 6))

for i in range(len(spectral_types)):
    intensity = planck_function(frequency_range, temperatures[i])
    plt.plot(np.log(frequency_range), np.log(intensity), label=f'Spectral Type {spectral_types[i]}(≈{temperatures[i]}K)')

# Marking the location of 1 Rydberg with a dashed vertical line
rydberg_frequency = 3.28984e15  # Hz
plt.axvline(x=np.log(rydberg_frequency), color='k', linestyle='dashdot', label='1 Rydberg')

# Adding labels and legend
plt.xlabel('Log(Frequency) (Hz)')
plt.ylabel('Log(Planck Function) (Joule s⁻¹ m⁻² sr⁻¹ Hz⁻¹)')
plt.title('Synthetic Planck Curves for Different Spectral Types')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord, Galactic
import astropy.units as u

# Load the stellar data
stellar_data = np.genfromtxt('obcat.dat', dtype=None, names=['l', 'b'], skip_header=0)  # Adjust skip_header as needed

# Extract Galactic coordinates of stars
galactic_coords = SkyCoord(l=stellar_data['l'], b=stellar_data['b'], frame='galactic')

# Extract coordinates of LMC, SMC, and Andromeda
lmc_coords = SkyCoord.from_name('LMC')
smc_coords = SkyCoord.from_name('SMC')
andromeda_coords = SkyCoord.from_name('Andromeda')

# Convert Galactic coordinates to Hammer-Aitoff projection
ra = galactic_coords.icrs.ra.wrap_at(180 * u.deg).radian
dec = galactic_coords.icrs.dec.radian

# Convert LMC, SMC, Andromeda coordinates to Hammer-Aitoff projection
lmc_ra = lmc_coords.icrs.ra.wrap_at(180 * u.deg).radian
lmc_dec = lmc_coords.icrs.dec.radian

smc_ra = smc_coords.icrs.ra.wrap_at(180 * u.deg).radian
smc_dec = smc_coords.icrs.dec.radian

andromeda_ra = andromeda_coords.icrs.ra.wrap_at(180 * u.deg).radian
andromeda_dec = andromeda_coords.icrs.dec.radian

# Plot the Hammer-Aitoff projection
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='hammer')

# Plot stars
ax.scatter(ra, dec, s=10, label='Bright Stars (O and B)', color='blue')

# Plot LMC, SMC, and Andromeda
ax.scatter(lmc_ra, lmc_dec, s=150, label='LMC', color='red', marker='o', alpha=0.7)
ax.scatter(smc_ra, smc_dec, s=150, label='SMC', color='green', marker='o', alpha=0.7)
ax.scatter(andromeda_ra, andromeda_dec, s=150, label='Andromeda', color='orange', marker='o', alpha=0.7)

# Add labels for external galaxies
ax.text(lmc_ra, lmc_dec, '  LMC', color='red', fontsize=10)
ax.text(smc_ra, smc_dec, '  SMC', color='green', fontsize=10)
ax.text(andromeda_ra, andromeda_dec, '  Andromeda', color='orange', fontsize=10)

# Add legend
ax.legend(loc='upper right')

# Add title and labels
plt.title('Distribution of Bright Stars and Nearby Galaxies in Galactic Coordinates')
plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')

plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# constants
h = 6.62607015e-34
c = 299792458.0
k = 1.380649e-23

# Initializing the lists
spectral_types = ['O', 'B', 'A', 'F', 'G', 'K', 'M']
temperatures = [30000, 25000, 10000, 7000, 5500, 4000, 3000]
radius = [15, 10, 2, 1, 0.8, 0.5, 0.08]  
Flux = []  

# Frequency range (in Hz)
frequency_range = np.linspace(3.29e15, 2.418e17, 100000)

# Defining the function for the Hydrogen Ionizing Flux in the units of Photon per second
def Ionizing_Flux(frequency, temperature, radius):
    return (((2 * frequency**2) * (4 * np.pi * (radius * (6.957e08))**2) * (np.pi)) / ((c**2) * (np.exp((h * frequency) / (k * temperature)) - 1))) 

# Calculating the Hydrogen Ionizing Flux for each spectral type by integrating using the trapezoid function
for i in range(len(spectral_types)):
    Unit_Frequency = Ionizing_Flux(frequency_range, temperatures[i], radius[i])
    Flux.append(np.trapz(Unit_Frequency, frequency_range))  

print(Flux)
```

```python
# Now finding the Stromgen Radius
stromgen_radius = []

# alpha_H is the recombination coefficient
alpha_H = 3 * pow(10, -13)

# Number density of electrons and protons is equal here
n = 120

# Finding the Stromgen Radius for each spectral type in pc
for i in range(len(spectral_types)):
    stromgen_radius.append((pow(((3 * Flux[i]) / (4 * np.pi * alpha_H *  pow(n,2))), 1/3)) * 3.2408 * pow(10, -19))

print(stromgen_radius)
```

```python
# plotting the graph for size of HII region vs spectral type
plt.bar(spectral_types, stromgen_radius)
plt.title('HII Region Sizes by Spectral Type')
plt.xlabel('Spectral Type')
plt.ylabel('HII Region Size (pc)')
plt.show()
```
